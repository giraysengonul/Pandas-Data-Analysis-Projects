```python
import pandas as pd
```

```python
movies = pd.read_csv(r'C:\Users\Giray\Downloads\imdb_movies.csv')
movies_data = movies
movies_data.head()
```

```python
pd.set_option('display.max.rows', 11000)
pd.set_option('display.max.columns',22)
```

```python
movies_data.drop_duplicates(inplace=True)
```

```python
movies_data[movies_data.duplicated()]
```

```python
movies_data.dropna(subset=['genres'],inplace=True)
```

```python
movies_data['profit'] = movies_data['revenue'] - movies_data['budget']
```

```python
movies_genre = movies_data[['popularity','budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average','profit']]
```

```python
movies_genre.head()
```

```python
from pandas import Series,DataFrame
```

```python
split_data = movies_genre['genres'].str.split('|').apply(Series,1).stack()
split_data.index = split_data.index.droplevel(-1)
split_data.name = 'genres_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split_data)
```

```python
movies_genre.head()
```

## Research Questions(Q):
#### 1.Which genres are the most common (number of movies made)?

```python
genres_count = pd.DataFrame(movies_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title',ascending=True)
```

```python
genres_count['original_title'].plot.pie(title='Movies Per genre in %',autopct='%1.1f%%',figsize=(10, 10))
```

```python
genres_count['original_title'].plot.barh(title='Movies Per Genre',color='DarkBlue',figsize=(10,9))
```

#### 2. Which genres have high avg. budget and revenue?

```python
movies_genre.head()
```

```python
genres_avg = movies_genre.groupby('genres_split').mean(numeric_only=True)
genres_avg.sort_values(by='budget',ascending=True,inplace=True)
pd.options.display.float_format = '{:,.2f}'.format
genres_avg
```

```python
genres_avg[['budget','revenue']].plot.barh(title= 'Budget and Revenue by Genre', color=('DarkBlue','c'),figsize=(10,9))
```

#### 3. Which genres have high avg. Profit?

```python
genres_avg.sort_values(by='profit',ascending= True, inplace=True)
genres_avg['profit'].plot.barh(title='Profit by Genre',color='DarkBlue',figsize=(10,9))
```

#### 4. Which genres have hight avg. popularity?

```python
genres_avg.sort_values(by='popularity',ascending= True, inplace=True)
genres_avg['popularity'].plot.barh(title='Popularity by Genre',color='DarkBlue',figsize=(10,9))
```

#### 5. Which genres have highest number of movies with an voting avg. >=8?

```python
movies_genre.head()
```

```python
vote_fifty= movies_genre[(movies_genre['vote_count']>=50) & (movies_genre['vote_average']>=8)]
vote_average = pd.DataFrame(vote_fifty.groupby('genres_split').vote_average.nunique()).sort_values(by='vote_average',ascending=True)
vote_average
```

```python
vote_average.plot.barh(title='Vote Average by Genre',color='DarkBlue',figsize=(10,9))
```

### Research Hypotheses (H):
1. The best movies according to vote avg. return high profit and revenue.


```python
movies_data.drop_duplicates(inplace=True)
movies_data.dropna(subset=['genres'],inplace=True)
movies_data['profit'] = movies_data['revenue'] - movies_data['budget']
movies_genre_2 = movies_data[['popularity','budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average','profit']]
movies_counted = movies_genre_2[movies_genre['vote_count']>= 50]
```

```python
movies_counted.corr(numeric_only=True,method='spearman')
```

```python
import seaborn as sea
import matplotlib.pyplot as plt
```

```python
sea.regplot(data=movies_counted,x='vote_average',y='profit',line_kws={'color':'red'})
#sea.regplot(data=movies_counted,x='vote_average',y='revenue',line_kws={'color':'green'})
```

2. The best movies according to popularity return high profit and revenue.

```python
sea.regplot(data=movies_counted,x='popularity',y='revenue',line_kws={'color':'red'})
plt.figure(figsize=(10,5))
plt.show()
```

3. Highly budgeted movies return high revenue and profit.

```python
sea.regplot(data=movies_counted,x='budget',y='profit',line_kws={'color':'red'})
plt.figure(figsize=(10,5))
plt.show()
```

4. Highly budgeted movies have a high popularity.

```python
sea.regplot(data=movies_counted,x='budget',y='popularity',line_kws={'color':'red'})
plt.figure(figsize=(10,5))
plt.show()
```

5. Look at Profit Per Genre Per Year

```python
time_genre = pd.DataFrame(movies_genre.groupby(['release_year','genres_split'])['profit'].mean())
final_genre = pd.pivot_table(time_genre, values='profit',index='genres_split',columns=['release_year'])
```

```python
sea.set(rc={'figure.figsize':(15,10)})
sea.heatmap(final_genre,cmap='YlGnBu',linewidths=0.5)
```
